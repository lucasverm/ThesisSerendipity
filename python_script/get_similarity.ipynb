```python
from gensim.models import KeyedVectors
if 'model' not in locals():
    model = KeyedVectors.load_word2vec_format('./GoogleNews-vectors-negative300.bin', binary=True)
```

```python
import pandas as pd
import numpy as np
from gensim.models import KeyedVectors

# Create empty dataframe for storing correlations
df = pd.DataFrame({'categorie': categories})
corr_df = pd.DataFrame(index=df.index, columns=df.index)
corr_df
# Fill in correlations
for i in range(len(df)):
    for j in range(i, len(df)):
        try:
            correlation = model.similarity(df.iloc[i]['categorie'], df.iloc[j]['categorie'])
        except:
            correlation = 0
        corr_df.iloc[i, j] = correlation
        corr_df.iloc[j, i] = correlation

# Replace NaN values with 0
corr_df = corr_df.fillna(0)
corr_df = corr_df.applymap(lambda x: round(x,4))
corr_df = corr_df.applymap(lambda x: 0 if x < 0 else x)
corr_df = corr_df.applymap(lambda x: 0 if x >= 1 else x)

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# create the MinMaxScaler object
scaler = MinMaxScaler()

# fit and transform the data
corr_df = scaler.fit_transform(corr_df)

# create a new DataFrame with the scaled values
dataframe = pd.DataFrame(corr_df)
dataframe.columns = categories
dataframe.index = categories
dataframe
```

```python
dataframe.to_csv('categorical_similarities.csv', index=True, sep=';',decimal=',')
```
