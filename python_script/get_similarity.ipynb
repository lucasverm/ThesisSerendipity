```python
import pandas as pd
import numpy as np
import difflib
from tqdm import tqdm
from time import sleep

col_names = row_names
df = pd.DataFrame(np.zeros((len(row_names), len(col_names))),
                  index=row_names, columns=col_names)
df

```

```python
from gensim.models import KeyedVectors
import pandas as pd

# Load pre-trained word embedding model
model = KeyedVectors.load_word2vec_format('./GoogleNews-vectors-negative300.bin', binary=True)

# Create a new dataframe with the same column and row labels as the original dataframe
sim_df = pd.DataFrame(columns=df.columns, index=df.columns)

# Loop through each column and row label and calculate the cosine similarity
for i in tqdm(row_names):
    for row_label in df.index:
        col_vector = model.get_vector(row_names[i])
        row_vector = model.get_vector(row_label)
        score = model.cosine_similarities(
            col_vector.reshape(1, -1), [row_vector])[0]
        sim_df.loc[row_label, row_names[i]] = score

print(sim_df)
```
