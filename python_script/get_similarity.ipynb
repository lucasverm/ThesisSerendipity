```python
if 'model' not in locals():
    model = KeyedVectors.load_word2vec_format(
        './GoogleNews-vectors-negative300.bin', binary=True)

# Create empty dataframe for storing correlations
df = pd.DataFrame({'categorie': categories})
corr_df = pd.DataFrame(index=df.index, columns=df.index)
# Fill in correlations
for i in range(len(df)):
    for j in range(i, len(df)):
        try:
            correlation = model.similarity(
                df.iloc[i]['categorie'], df.iloc[j]['categorie'])
            if df.iloc[i]['categorie'] == df.iloc[j]['categorie']:
                correlation = 0
        except:
            correlation = 0
        corr_df.iloc[i, j] = correlation
        corr_df.iloc[j, i] = correlation

# Replace NaN values with 0
corr_df = corr_df.fillna(0)
corr_df = corr_df.applymap(lambda x: 0 if x < 0 else x)
corr_df = corr_df.applymap(lambda x: 1 if x >= 1 else x)

# create the MinMaxScaler object
scaler = MinMaxScaler()
scaled_array = scaler.fit_transform(corr_df)
dataframe = pd.DataFrame(scaled_array)
dataframe = dataframe.applymap(lambda x: round(x, 4))
dataframe.columns = categories
dataframe.index = categories

#diagonaal = 1
identity = np.identity(dataframe.shape[0])
dataframe[identity == 1] = 1

dataframe
```
